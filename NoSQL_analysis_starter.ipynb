### 1. Which establishments have a hygiene score equal to 20?

```python
establishments.find_one()
```

```python
# Find the establishments with a hygiene score of 20
query1 = {'scores.Hygiene' : 20}
results1 = establishments.find(query1)
# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query1))
# Display the first document in the results using pprint
pprint(results1[0])
```

```python
# Convert the result to a Pandas DataFrame
df1 = pd.DataFrame(results1)
# Display the number of rows in the DataFrame
print(len(df1))
# Display the first 10 rows of the DataFrame
df1.head()
```

```python
df1.to_csv('Resources/hygiene20.csv', index=False)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
establishments.find_one({'LocalAuthorityName': {'$regex':'London'}})
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName':{'$regex':'London'}, 'RatingValue': {'$gte': 4}}
results2 = establishments.find(query2)
# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query2))
# Display the first document in the results using pprint
pprint(results2[0])
```

```python
# Convert the result to a Pandas DataFrame
df2 = pd.DataFrame(results2)
# Display the number of rows in the DataFrame
print(len(df2))
# Display the first 10 rows of the DataFrame
df2.head()
```

```python
df2.to_csv('Resources/London4andUp.csv', index=False)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
pprint(establishments.find_one({'BusinessName':'Penang Flavours'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query3 = {'geocode.latitude' : {'$gte' : latitude - degree_search},
         'geocode.latitude' : {'$lte' : latitude + degree_search},
         'geocode.longitude' : {'$gte' : longitude - degree_search},
         'geocode.longitude' : {'$lte' : longitude + degree_search},
         'RatingValue' : 5
         }
sort3 = [('scores.Hygiene',1)]
limit3 = 10 # only 10 so it isn't too cluttered
results3 = establishments.find(query3).sort(sort3).limit(limit3)
# Print the results   # this messes with my exports
#for obj in results3:
#    pprint(obj)
```

```python
# Convert result to Pandas DataFrame
df3 = pd.DataFrame(results3)
df3.head()
```

```python
df3.to_csv('Resources/Rating5nearPenang.csv', index=False)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
matchq = {'$match' : {'scores.Hygiene' : 0}}
groupq = {'$group' : {
    '_id' : '$LocalAuthorityName',
    'count' : {'$sum' : 1}}}
sortq = {'$sort' : {'count' : -1}}
pipeline = [matchq,groupq,sortq]
resultsagg = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
len(resultsagg)
# Print the first 10 results
resultsagg[0:10]
```

```python
# Convert the result to a Pandas DataFrame
df4 = pd.DataFrame(resultsagg)
# Display the number of rows in the DataFrame
print(len(df4))
# Display the first 10 rows of the DataFrame
df4.head()
```

```python
df4.to_csv('Resources/LowHygPerTown.csv', index=False)
```
